## Rolling correlation- along with multiIndexes and groupby

```python
ts_corr=ind_return.rolling(window=36).corr()
```

```python
ts_corr.tail()
```

```python
# we want to find the avarage corr
```

```python
ts_corr.index.names
```

```python
ts_corr.index.names=['date','industry']
```

```python
ts_corr.tail()
```

```python
ind_tr36corr=ts_corr.groupby(level='date').apply(lambda cormat:cormat.values.mean())
ind_tr36corr.plot()
```

```python
tmi_tr36rets.plot(label="tr36 mo rets",legend=True)
ind_tr36corr.plot(label="tr36 mo corr",legend=True,secondary_y=True)
```

```python
# we see that when market is falling the corr is raising and when rising they are falling
```

```python
tmi_tr36rets['2007':].plot(label="tr36 mo rets",legend=True)
ind_tr36corr['2007':].plot(label="tr36 mo corr",legend=True,secondary_y=True)
```
